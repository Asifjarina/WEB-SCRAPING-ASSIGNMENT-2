```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

url = "https://www.shine.com/job-search/data-analyst-jobs-in-bangalore"
page = requests.get(url)

soup = BeautifulSoup(page.content, "html.parser")
results = soup.find_all("li", class_="search_listing")

job_titles = []
job_locations = []
company_names = []
experience_required = []

for result in results[:10]:
    title_elem = result.find("h3", class_="job_title")
    location_elem = result.find("span", class_="job_loc")
    company_elem = result.find("a", class_="company_name")
    experience_elem = result.find("span", class_="exp")

    job_titles.append(title_elem.text.strip())
    job_locations.append(location_elem.text.strip())
    company_names.append(company_elem.text.strip())
    experience_required.append(experience_elem.text.strip())

data = {
    "Job Title": job_titles,
    "Job Location": job_locations,
    "Company Name": company_names,
    "Experience Required": experience_required,
}

df = pd.DataFrame(data)
print(df)
```

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

url = "https://www.azquotes.com/top_quotes.html"
response = requests.get(url)
soup = BeautifulSoup(response.content, "html.parser")

quotes = []
authors = []
types = []

for quote in soup.find_all("div", class_="wrap-blockquote"):
    quotes.append(quote.find("a", class_="title").text.strip())
    authors.append(quote.find("a", class_="author").text.strip())
    types.append(quote.find("a", class_="kw").text.strip())

df = pd.DataFrame({"Quote": quotes, "Author": authors, "Type of Quote": types})
print(df)
```

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

url = "https://www.jagranjosh.com/general-knowledge/list-of-all-prime-ministers-of-india-1473165149-1"
page = requests.get(url)

soup = BeautifulSoup(page.content, "html.parser")
results = soup.find_all("table", class_="table4")

name_list = []
born_dead_list = []
term_of_office_list = []
remarks_list = []

for result in results:
    rows = result.find_all("tr")
    for row in rows[1:]:
        cols = row.find_all("td")
        name = cols[0].text.strip()
        born_dead = cols[1].text.strip()
        term_of_office = cols[2].text.strip()
        remarks = cols[3].text.strip()

        name_list.append(name)
        born_dead_list.append(born_dead)
        term_of_office_list.append(term_of_office)
        remarks_list.append(remarks)

data = {
    "Name": name_list,
    "Born-Dead": born_dead_list,
    "Term of Office": term_of_office_list,
    "Remarks": remarks_list,
}

df = pd.DataFrame(data)
print(df)
```

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

url = "https://www.motor1.com/features/308149/most-expensive-new-cars-ever/"
page = requests.get(url)

soup = BeautifulSoup(page.content, "html.parser")
results = soup.find_all("div", class_="slide-item")

car_names = []
car_prices = []

for result in results:
    name_elem = result.find("h2", class_="slide-title")
    price_elem = result.find("div", class_="slide-price")

    car_names.append(name_elem.text.strip())
    car_prices.append(price_elem.text.strip())

data = {
    "Car Name": car_names,
    "Price": car_prices,
}

df = pd.DataFrame(data)
print(df)
```
